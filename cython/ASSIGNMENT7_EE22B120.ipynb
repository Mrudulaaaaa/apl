```python
#generating matrices
u = np.random.random((10,10))
v = np.random.random((10,10))
```

```python
def matrix_multiply(u, v):
    m, n = u.shape
    n, p = v.shape
    res = np.zeros((m, p))
    for i in range(m):
        for j in range(p):
            res[i,j] = 0
            for k in range(n):
                res[i,j] += u[i,k] * v[k,j]
    return res
```

```python
%timeit matrix_multiply(u,v)
```

```python
%timeit (u@v)
```

- There's no much change in changing th type from double to float. The time remains same.

We can conclude that using cython in defining functions with several transformations:
- Loading the cython extension, and importing cython to the matrix_multiply i.e `cy_matrix_multiply` decreased the time taken by approximately 10msec.
- Using the decorator:
    - @cython.boundscheck(False) - Avoids checking the bounds of the arrays. 
    - It assumes indices are always valid.
    - It reduced the error time from milli seconds to micro seconds. Not a significant change.
- Declaring the type of input varibales:
    - The time decreased by approximately half the initial value.
    - It avoids typechecks of input.
- Declaring the types of both input, output varibales:  
    - There is significant decrease in time. It was initially in milli seconds and now it decreased to micro seconds. The computational time has decreased by good amount i.e 10^-3 times.
    - It avoids typechecks during compilation, as cython compiler knows the datatypes before hand.
    - It can optimize memory layout more effectively when variable types are declared, which can reduce memory access latency.
    - This generates a more efficient code, hence significant decrease in compuatational time.
    
The best change is observed when the types of input, output varibaes are declared. The time decreased when compared with `matrix_multiply` before. But, `numpy matrix multiplication` still wins. The time is in 10's of microseconds for numpy multiplication and in 100's for `cy_matrix_multiply`.
